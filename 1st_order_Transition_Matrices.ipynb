```python
import numpy as np
import pandas as pd
```

```python
char_dict = {
    ' ':0,   
    'a':1,
    'b':2,
    'c':3,
    'd':4,
    'e':5,
    'f':6,
    'g':7,
    'h':8,
    'i':9,
    'j':10,
    'k':11,
    'l':12,
    'm':13,
    'n':14,
    'o':15,
    'p':16,
    'q':17,
    'r':18,
    's':19,
    't':20,
    'u':21,
    'v':22,
    'w':23,
    'x':24,
    'y':25,
    'z':26,
    'á':27,
    'é':28,
    'í':29,
    'ó':30,
    'ú':31,
    'ñ':32,
    'ü':33,
    'ä':34,
    'å':35,
    'ö':36,
    'ß':37,
}

size = len(char_dict)
size
```

```python
def updateT(word, char, T):
    
    if char in word:
        
        row = char_dict[char]
        indice = [i for i, letter in enumerate(word) if letter == char]
        count = len(indice)
        #print('count=%s'%count)
        T[row][-1] += count
        col=None
        
        next_indice = [i+1 for i in indice]
        #print('next_indice=%s'%next_indice)
        
        for i in next_indice:
            
            if i == len(word):
                T[row][0] += 1
            else:
                next_char = word[i]
                #print('next_char=%s'%next_char)
                col = char_dict[next_char]
                #print(row)
                #print('updated col=%s'%col)
                T[row][col] += 1
    
    else:
        None
    
```

```python
def computeT(df, T):
    rows, columns = df.shape[0], df.shape[1]
    for char in char_dict:
        #print('char=%s'%char)
        for r in range(rows):
            #print('r = %s'%r)
            word = df.iloc[r][0]
            updateT(word, char, T)
```

## English

```python
df_E = pd.read_csv("1000 most common words - English.csv", header=None)

# Data cleaning
df_E = df_E.replace(to_replace="don't",value='dont')
df_E.iloc[241,0] = 'dont'
df_E.iloc[751,0] = 'wont'

T_E = np.zeros((size,size+1))

computeT(df_E, T_E)
```

```python
T_E_counts = T_E
```

```python
T_E_counts
```

```python
df_TE_C = pd.DataFrame(T_E_counts[:,:-1], index=char_dict, columns = char_dict, dtype=int)
df_TE_C.head()
```

```python
total = T_E[:,-1]
for i in range(T_E.shape[0]):
    if total[i] != 0:
        T_E[i] = T_E[i] / total[i]
T_E = T_E[:,:-1]
```

```python
T_E
```

```python
df_TE = pd.DataFrame(T_E, index=char_dict, columns = char_dict)
df_TE.head()
```
