```python
import pandas as pd
import numpy as np
import csv
```

```python
df = pd.read_csv('1000 most common words - Train.csv', header = None)
df.columns = ["Word"]
df = df["Word"].str.split(r'', expand=True)
df = df.replace('', 0)
df = df.fillna(value=0)
df.drop(df.columns[[0,17]],axis=1,inplace=True)

for i in range(1,17):
    if 0 in df[i]:
        continue
    else:
        df[i] = df[i].str.lower()
df.to_csv('Separated.csv',header=False,index=False)
df
```
